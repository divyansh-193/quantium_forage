Stores with store numbers 77, 86 and 88 are trial stores. We will now examine its performance compared to the non-trial stores.

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Retrieving previous data combined from Task 1

```python
sales_data = pd.read_csv("/content/drive/MyDrive/out_sales_data.csv")
sales_data.head()
```

```python
sales_data
```

```python
sales_data.describe()
```

Grouping all the trial stores into one dataset for comparing them with the non-trial ones.

```python
trial_stores = sales_data[sales_data['STORE_NBR'].isin([77, 86, 88])]
trial_stores.head()
```

Summary statisitcs for the Trial stores

```python
trial_stores.describe()
```

```python
trial_stores['TOT_SALES'].sum()
```

```python
trial_stores['STORE_NBR'].count()
```

```python
trial_stores['TOT_SALES'].mean()
```

```python
transactions_per_card = trial_stores.groupby('LYLTY_CARD_NBR')['TOT_SALES'].count().mean()
print(f"Average number of transactions per unique LYLTY_CARD_NBR for the trial stores: {transactions_per_card}")
```

Formed dataset of Non-trial stores by combining Trial stores and Overall dataset and dropped the duplicates (which was Trial stores data)

```python
nontrial_stores = pd.concat([sales_data, trial_stores]).drop_duplicates(keep=False)
nontrial_stores.head()
```

Summary statistics of the stores which were under trail, functioning normally as before

```python
nontrial_stores.describe()
```

```python
nontrial_stores['TOT_SALES'].sum()
```

```python
nontrial_stores['TOT_SALES'].count()
```

```python
nontrial_stores['TOT_SALES'].mean()
```

```python
transactions_per_cardn = nontrial_stores.groupby('LYLTY_CARD_NBR')['TOT_SALES'].count().mean()
print(f"Average number of transactions per unique LYLTY_CARD_NBR for the non trial stores: {transactions_per_cardn}")
```

As we can see Trial stores have more of average sales than Non-trial ones by \$0.3 per unique Loyalty card number. And we can see similar trend in the mean of transactions, trial stores having $0.2 more on each transaction compared to non-trial stores.
